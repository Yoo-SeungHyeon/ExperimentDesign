# MLB 1990년대 분석

## 상관분석

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('MLB_1990.csv')
data
```

```python
data['win']
```

```python
WA = list(data['win'])
BRG = list(data['득점'])
PRG = list(data['실점'])
```

```python
BDP = []
for i in range(278):
    BDP.append(BRG[i]/PRG[i])
```

```python
plt.subplot(221)
plt.scatter(WA,BRG,alpha = 0.5, color = 'gray')
plt.subplot(222)
plt.scatter(WA,PRG,alpha = 0.5)
plt.subplot(223)
plt.scatter(WA,BDP,alpha = 0.5, color = 'green')
plt.savefig('Correl_Analy.png')
```

```python
plt.scatter(WA,BRG,alpha = 0.5, color = 'gray')
plt.savefig('CA_WA_BRG.png')
```

```python
plt.scatter(WA,PRG,alpha = 0.5)
plt.savefig('CA_WA_PRG.png')
```

```python
plt.scatter(WA,BDP,alpha = 0.5,color = 'green')
plt.savefig('CA_WA_BDP.png')
```

```python
import scipy.stats as stats
```

```python
display(stats.pearsonr(WA,BRG),
stats.pearsonr(WA,PRG),
stats.pearsonr(WA,BDP),
stats.pearsonr(BRG,PRG),
stats.pearsonr(BRG,BDP),
stats.pearsonr(PRG,BDP))
```

```python
print('WA 와 BRG 상관분석')
print('Correlation : {:.4f}'.format(stats.pearsonr(WA,BRG)[0]))
print('P-value : {:.4f}'.format(stats.pearsonr(WA,BRG)[1]))
print('-----------------------------------------------------')
print('WA 와 PRG 상관분석')
print('Correlation : {:.4f}'.format(stats.pearsonr(WA,PRG)[0]))
print('P-value : {:.4f}'.format(stats.pearsonr(WA,PRG)[1]))
print('-----------------------------------------------------')
print('WA 와 BRG/PRG 상관분석')
print('Correlation : {:.4f}'.format(stats.pearsonr(WA,BDP)[0]))
print('P-value : {:.4f}'.format(stats.pearsonr(WA,BDP)[1]))
print('-----------------------------------------------------')
print('BRG 와 PRG 상관분석')
print('Correlation : {:.4f}'.format(stats.pearsonr(BRG,PRG)[0]))
print('P-value : {:.4f}'.format(stats.pearsonr(BRG,PRG)[1]))
print('-----------------------------------------------------')
print('BRG 와 BRG/PRG 상관분석')
print('Correlation : {:.4f}'.format(stats.pearsonr(BRG,BDP)[0]))
print('P-value : {:.4f}'.format(stats.pearsonr(BRG,BDP)[1]))
print('-----------------------------------------------------')
print('PRG 와 BRG/PRG 상관분석')
print('Correlation : {:.4f}'.format(stats.pearsonr(PRG,BDP)[0]))
print('P-value : {:.4f}'.format(stats.pearsonr(PRG,BDP)[1]))
```

## 회귀분석

```python
import statsmodels.api as sm
```

```python
DataSet = pd.DataFrame({'WA' : WA,
                       'BRG' : BRG,
                       'PRG' : PRG,
                       'BRG/PRG' : BDP})
```

### 중선형 회귀 분석 WA & BRG, PRG

```python
x_data = DataSet[["BRG","PRG"]]
target = DataSet[["WA"]]
x_data1 = sm.add_constant(x_data, has_constant = "add")

multi_model = sm.OLS(target, x_data1)
fitted_multi_model = multi_model.fit()
fitted_multi_model.summary()
```

### 단순 선형 WA & BRG/PRG

```python
lr_X = sm.add_constant(BDP, has_constant = "add")
lr = sm.OLS(WA,lr_X)
result = lr.fit()
result.summary()
```
